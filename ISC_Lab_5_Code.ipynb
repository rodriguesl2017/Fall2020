# Logistic Map Part 2: Poincaré Plots, Bifurcation Diagrams, and Cobweb Plots 

```
import numpy as np
import matplotlib.pyplot as plt #package import
```

```
def logistic(r, x):
    return r * x * (1 - x)  #logistic definition
```

```
n = 100
r = np.linspace(2.5, 4.0, n) #creates vector r with n values equally spaced between 2.5 and 4
```

```
x = 1e-5 * np.ones(n)  #new vector with n 1e-5 values
```

```
iterations = 150 #sets how many iterations
```

```
x = logistic(r, x)
```

```
X = np.zeros((iterations,n)) #creates iterations x n matrix
```

```
for i in range(1,iterations):#fills first column of matrix with each iteration of the logistic function
    
    x = logistic(r, x) 

    X[i,:] = x
```

```
plt.plot(X[:,[1]]) #plots the first column of the matrix
```

```
plt.plot(X[:,80],'.') #plots a further iteration
```

# Bifurcation Diagrams

```
n = 100
r = np.linspace(2.5, 4.0, n) 
```

```
x = 1e-5 * np.ones(n) 
```

```
fig, ax1 = plt.subplots(1, 1, figsize=(10,10))
ax1.plot(r, x, '.', alpha=.25,markersize=10) #plots the r and x
```

```
iterations = 150
```

```
x = logistic(r, x) #apply logistic function to x values
```

```
fig, ax1 = plt.subplots(1, 1, figsize=(10,10)) #plots r vs the logistic x values
ax1.plot(r, x, '.', alpha=.25, markersize=10)
```

```
fig, ax1 = plt.subplots(1, 1, figsize=(10,10)) #iterates the above graph using the logistic map

for i in range(1,iterations):

    x = logistic(r, x)

    ax1.plot(r, x, '.b', alpha = (i + 1) / iterations)
```

# Cobweb Plots 

<img src='https://upload.wikimedia.org/wikipedia/commons/e/ed/LogisticCobwebChaos.gif'>

An animated cobweb diagram of the logistic map $y = r x (1-x)$, showing chaotic behaviour for most values of $r > 3.57$.

```
x = np.linspace(0, 1) #vector with values between 0 and 1 equally spaced out
fig, ax = plt.subplots(1, 1) 
ax.plot(x, logistic(2, x), 'k') #iterates values once and plot it
```

```
def plot_system(r, x0, n): #function definition for the cobweb plot

    
    t = np.linspace(0, 1) #vector with values between 0 and 1 equally spaced out

    fig, ax = plt.subplots(1, 1, figsize=(10, 10)) #plots 1 iteration
    ax.plot(t, logistic(r, t), 'k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)

    x = x0 #point we are iterating around
    for i in range(n):
        
        y = logistic(r, x) #each iteration assings a new y value

        ax.plot([x, x], [x, y], 'k', lw=1)
        ax.plot([x, y], [y, y], 'k', lw=1)
        ax.plot([x], [y], 'ok', ms=10, alpha=(i + 1) / n) #cobwebbing

        x = y

    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title(f"$r={r:.1f}, \, x_0={x0:.1f}$")
```

```
plot_system(3.6, .25, 10) #plots using the defined function
```

```
plot_system(3.2, .1, 10) #other parameters
```

```
plot_system(3.9, .1, 10) #other parameters
```
