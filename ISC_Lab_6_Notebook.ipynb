```
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt
%matplotlib inline #import packages
```

```
m = 1.  # m for mass
k = 0.  # k for drag
g = 9.81  # gravity force in m/s^2
```

```
# stores informaiton for trajectory suhc as velocity and acceleration
v0 = np.zeros(4)
v0[2] = 4.   
v0[3] = 10.
```

```
def f(v, t0, k):
    u, udot = v[:2], v[2:]  #storing position and velocity

    udotdot = -k / m * udot #drag coefficient is added to forces
    udotdot[1] -= g #net acceleration
    
    return np.r_[udot, udotdot]
```

```
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

t = np.linspace(0., 3., 30)

for k in np.linspace(0., 1., 5): 

    v = spi.odeint(f, v0, t, args=(k,)) #iteratively solves diffenetial equatiosn

    ax.plot(v[:, 0], v[:, 1], 'o-', label=f'k={k:.1f}') #plot each
    
ax.legend()
ax.set_xlim(0, 12)  
ax.set_ylim(0, 6)
```
